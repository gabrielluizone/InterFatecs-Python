```python
import pandas as pd
```

```python
def main():
  n = int(input())
  city = [['ðŸ…²' for _ in range(n)] for _ in range(n)]

  for idx in range(1, int(n/2)+1, 2):
    for r in range(idx, n-idx): # Borda superior
      city[idx][r] = 'R'

    for r in range(idx, n-idx): # Borda inferior
      city[n-idx-1][r] = 'R'

    for r in range(idx, n-idx): # Esquerda
      city[r][idx] = 'R'

    for r in range(idx, n-idx): # Direita
      city[r][n-idx-1] = 'R'

  # Adicionando 'A' na diagnoal principal
  for a in range(n): city[a][a] = 'A'
  display(pd.DataFrame(city))

  return print(sum(rua.count('ðŸ…²') for rua in city))

if __name__ == '__main__':
  main()
```
